```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/data.csv')
df['Gender'] == 'Female'
```

```python
fdata=df[df['Gender'] == "Female"]
print("THE FEMALE EMPLOYEES LEFT THE COMPANY ")
fdata
```

```python
mdata=df[df['Gender'] == "Male"]
print("THE MALE EMPLOYEES LEFT THE COMPANY ")
mdata
```

```python
print("The following inference shows mean age of female and male Employees")
df[["Gender", "Age"]].groupby("Gender").mean()
```

```python
print("The following inference shows the mean method is applied to each column containing numerical columns by passing numeric_only=True:")
df.groupby("Gender").mean(numeric_only=True)
```

```python
gender = df['Gender'].value_counts()
gender
```

```python
import numpy as np
import matplotlib.pyplot as plt
print("from the below graph it is clear that  more male employees left the company than Female employees")
plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="c")
ax.set_title("Bar Graph of Gender", y = 1)
ax.set_xlabel('Gender')
ax.set_ylabel('Number of Employees')
ax.set_xticklabels(('Male', 'Female'))
```
